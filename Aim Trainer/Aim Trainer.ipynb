```python
from selenium import webdriver
import time
import numpy as np
import cv2
import win32api
import win32con
from ScreenGrabber import ScreenGrabber

driver = webdriver.Chrome()
page_source = "https://humanbenchmark.com/tests/aim"
driver.get(page_source)
time.sleep(2) 
#print("Page loaded successfully!")

# Initialize ScreenGrabber with specified coordinates
focus_area = ScreenGrabber(top_left_x=540, top_left_y_from_top=330, width=2000-540, height=1055-330)

def detect_white_circle_center(screen_shot):
    gray_image = cv2.cvtColor(screen_shot, cv2.COLOR_BGR2GRAY)
    _, white_pixels = cv2.threshold(gray_image, 240, 255, cv2.THRESH_BINARY)
    contours, _ = cv2.findContours(white_pixels, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if contours:
        largest_contour = max(contours, key=cv2.contourArea)
        M = cv2.moments(largest_contour)        
        # Calculate the center of the contour
        if M["m00"] != 0:
            cx = int(M["m10"] / M["m00"])
            cy = int(M["m01"] / M["m00"])
            return cx, cy
    return None, None

def click_target(x, y):
    win32api.SetCursorPos((x + 540, y + 330)) 
    win32api.mouse_event(win32con.MOUSEEVENTF_LEFTDOWN, 0, 0)
    win32api.mouse_event(win32con.MOUSEEVENTF_LEFTUP, 0, 0)

for _ in range(80):
    focus_area.take_screen_shot()
    screen_shot = focus_area.get_screenshot_segment(0, 0, 2000-540, 1055-330)

    # Detect the target and click
    target_x, target_y = detect_white_circle_center(screen_shot)
    if target_x is not None and target_y is not None:
        click_target(target_x, target_y)
        print(f"Clicked at ({target_x}, {target_y})")
    else:
        print("No target detected.")
    # time.sleep(0.1)
```
